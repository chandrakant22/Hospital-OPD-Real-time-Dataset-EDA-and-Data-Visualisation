```python
ds['Date']=pd.to_datetime(ds['Date'])
ds
```

```python
ds[['Day','Month','Year']] = ds['Date'].str.split("/|-",n=2,expand=True)
ds
```

```python
ds['Day']=pd.to_numeric(ds['Day'])
```

```python
a=ds['Day'].value_counts()
a.sort_index()
```

```python
for i in range (1,32,1):
  print('For date: ',str(i)+'-12-2022')
  m=ds.loc[ds['Day']== i].value_counts('Department name')
  print(m,'\n')
```

```python
a=ds.groupby(['Day', 'Department name'])['Day'].count().unstack('Department name').fillna(0)
a.plot(kind='bar',stacked=True)
```
